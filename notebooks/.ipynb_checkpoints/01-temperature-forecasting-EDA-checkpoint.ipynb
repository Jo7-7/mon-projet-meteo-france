```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# add low_memory=False to fix the warning
df = pd.read_csv('../data/raw/city_temperature.csv', low_memory=False)

# Check the DataFrame again clearly
print(df.head())
df.info()

print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
df['AvgTemperature'].hist(bins=50)
plt.title('Distribution of Average Temperature')
plt.xlabel('Avg Temperature')
plt.ylabel('Frequency')
plt.show()
```

```python
# clearly use .copy() here to fix the warning
city_df = df[(df['City'] == 'Paris') & (df['Country'] == 'France')].copy()

# Now explicitly create your Date column clearly
city_df.loc[:, 'Date'] = pd.to_datetime(city_df[['Year', 'Month', 'Day']])

# Sort by date clearly
city_df = city_df.sort_values('Date')

# Plot the temperature clearly
plt.figure(figsize=(12,6))
plt.plot(city_df['Date'], city_df['AvgTemperature'], linewidth=0.5)
plt.title('Temperature Trend in Paris')
plt.xlabel('Date')
plt.ylabel('Avg Temperature')
plt.grid()
plt.show()
```

```python
# Check explicitly for suspicious temperatures clearly
df[df['AvgTemperature'] == -99].head()
```

```python
# Removing the rows with invalid temperatures clearly (-99F)
df_cleaned = df[df['AvgTemperature'] != -99]
```

```python
df = df.drop('State', axis=1)
```

```python
# Check how many NaNs we now have clearly
df.isnull().sum()

# Decide clearly: you could drop these NaNs
df.dropna(subset=['AvgTemperature'], inplace=True)

# Clearly verify the final cleaned dataframe
df.info()
```
